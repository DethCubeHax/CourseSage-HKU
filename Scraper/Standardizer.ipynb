```python
import json
import os
import pandas as pd
data = []
DIR = "data/"
rawFacultyNames = []
```

```python
tmp = []

with open(f"{DIR}courseList.txt", "r") as f:
    tmp = f.read()
    tmp = tmp.split(",\n")
    tmp = tmp[:-1]

for item in tmp:
    data.append(json.loads(item))
```

```python
for item in data:
    if item["offer_dept"] not in rawFacultyNames:
        rawFacultyNames.append(item["offer_dept"])
```

```python
business = ["Faculty of Business and Econom", "Business & Economics Faculty"]

arts = ["Dept of Architecture", "School of Modern Languages & C", "Architecture Faculty", 
        "School of Humanities", "School of Chinese", "Dept of Real Estate & Construc",
        "Dept of Geography", "Arts Faculty", "Dept of Urban Planning & Desig"]

dentistry = []

education = ["Education Faculty"]

engineering = ["Engineering Faculty", "Dept of Civil Engineering", "Dept of Computer Science",
                "Dept of Electrical & Electroni", "Dept of Industrial & Mfg Sys E", "Dept of Mechanical Engg"]

law = ["Dept of Law", "Law Faculty", "Dept of Prof Legal Education"]

medicine = ["School of Biomedical Sciences", "School of Chinese Medicine", 
        "LKS Medical Faculty", "LKS Faculty of Medicine", "Dept of Pharmacology & Pharmac",
        "School of Public Health", "Dental Faculty", "Dept of Diagnostic Radiology,",
        "School of Clinical Medicine", "School of Clinical Medicine", "Dept of Medicine", 
        "Dept of Microbiology", "Dept of Pathology, SClinMed", "Dept of Pathology, SClinMed"
        ,"Dept of Orthop & Traumatol, SC", "Dept of Surgery, SClinMed", "Dept of Obstetrics & Gynaecolo",
        "Dept of Obstet & Gynaecol, SCl", "School of Nursing", "Dept of Orthopaedics & Traumat",
        "Dept of Paediatr & Adolesc Med", "Dept of Psychiatry, SClinMed", "Dept of Psychiatry, SClinMed"
        "Dept of Surgery"]

science = ["Science Faculty", "Dept of Mathematics", "School of Biological Sciences", 
        "Dept of Microbiology, SClinMed", "Dept of Physics"]

socSci = ["Social Sciences Faculty", "Centre of Buddhist Studies", "Centre for Civil Society & Gov",
            "Centre for the Humanities and", "Dept of Soc Work & Social Admi", "Dept of Earth Sciences"
            "HK Institute for Humanities &", "Dept of Politics & Public Admi", "Dept of Psychology"
            , "Dept of Psychiatry, SClinMed", "Journalism and Media Studies C"]

caes = ["Centre for Applied English Stu", "School of English"]

grad = ["Dept of Statistics & Actuarial", "Graduate School"]

hkuSpace = ["University Central"]


```

```python
for item in data:
    rawDept = item["offer_dept"]
    if rawDept in business:
        item["offer_dept"] = "Faculty of Business and Economics"
    elif rawDept in arts:
        item["offer_dept"] = "Faculty of Arts"
    elif rawDept in dentistry:
        item["offer_dept"] = "Faculty of Dentistry"
    elif rawDept in education:
        item["offer_dept"] = "Faculty of Education"
    elif rawDept in law:
        item["offer_dept"] = "Faculty of Law"
    elif rawDept in medicine:
        item["offer_dept"] = "Faculty of Medicine"
    elif rawDept in science:
        item["offer_dept"] = "Faculty of Science"
    elif rawDept in socSci:
        item["offer_dept"] = "Faculty of Social Sciences"
    elif rawDept in grad:
        item["offer_dept"] = "Graduate School"
    elif rawDept in hkuSpace:
        item["offer_dept"] = item["code"][:4]
    elif rawDept in caes:
        item["offer_dept"] = "Center for Applied English Studies"
    elif rawDept in engineering:
        item["offer_dept"] = "Faculty of Engineering"
    else:
        item["offer_dept"] = None
```

```python
facs = {}
for item in data:
    if item["offer_dept"] is not None:
        if item["code"][:4] not in facs:
            facs[item["code"][:4]] = item["offer_dept"]

for item in data:
    if item["offer_dept"] is None:
        if item["code"][:4] in facs:
            item["offer_dept"] = facs[item["code"][:4]]
```

```python
sortedFacultyNames = []
for item in data:
    if item["offer_dept"] not in sortedFacultyNames:
        sortedFacultyNames.append(item["offer_dept"])
```

```python
coursesByGrade = []

tmp = []

with open(f"{DIR}sortedGrades.txt", "r") as f:
    tmp = f.read()
    tmp = tmp.split(",\n")
    tmp = tmp[:-1]

for item in tmp:
    coursesByGrade.append(json.loads(item))
```

```python
gradesSortedByFaculty = []
for faculty in sortedFacultyNames:
    for course in coursesByGrade:
        courseName = course["code"]
        for item in data:
            if item["offer_dept"] is not None:
                if (item["code"] == courseName) and (item["offer_dept"]) == faculty:
                    gradesSortedByFaculty.append(item)
                    
```

```python
coursesByReview = []
tmp = []

with open(f"{DIR}sortedReviews.txt", "r") as f:
    tmp = f.read()
    tmp = tmp.split(",\n")
    tmp = tmp[:-1]

for item in tmp:
    coursesByReview.append(json.loads(item))
```

```python
reviewsSortedByFaculty = []
for faculty in sortedFacultyNames:
    for course in coursesByReview:
        courseName = course["code"]
        for item in data:
            if item["offer_dept"] is not None:
                if (item["code"] == courseName) and (item["offer_dept"]) == faculty:
                    reviewsSortedByFaculty.append(item)
```

```python
allData = []
tmp = []
instructorList = []
bestReviews = []

with open(f"{DIR}grades.txt", "r") as f:
    tmp = f.read()
    tmp = tmp.split(",\n")
    tmp = tmp[:-1]

for item in tmp:
    instructorList.append(json.loads(item))

for faculty in sortedFacultyNames:
    if faculty == "" or faculty == None: continue
    facultySchema = {
        "facultyName": faculty,
        "sortedByGrades": [],
        "sortedByReviews": []
    }
    for item in gradesSortedByFaculty:
        if item["offer_dept"] == faculty:
            instructors = []
            instructorsArray = []
            reviews = []
            for instructor in instructorList:
                if instructor["code"] == item["code"]:
                    if "instructors" in instructor:
                        instructors = instructor
            if os.path.isfile(f"{DIR}/results/{item['code']}.csv"):
                tmp = []
                with open(f"{DIR}/results/{item['code']}.csv", "r") as f:
                    tmp = f.read()
                    tmp = tmp.split("\n")
                    tmp = tmp[:-1]
                skipFirst = True
                for review in tmp:
                    if not skipFirst:
                        reviews.append(review.split(","))
                    skipFirst = False
            if len(instructors) > 0:
                for ins in instructors["instructors"]:
                    if ins["instructor"] not in instructorsArray:
                        if ins["instructor"] is not None:
                            instructorsArray.append(ins['instructor'])
            
            instructors = []
            reviews = []
            for instructor in instructorList:
                if instructor["code"] == item["code"]:
                    instructors = instructor
            
            courseforGradesSchema = {
                "courseCode": item["code"],
                "courseName": item["title"],
                "courseInstructor": instructorsArray,
                "bestReview": None,
                "gradeList": None,
                "gradeListDetailed": None
            }
            sentAnal = []
            skipFirst = True
            for review in tmp:
                if not skipFirst:
                    sentAnal.append(review.split(","))
                skipFirst = False
            if len(sentAnal) > 0:
                courseforGradesSchema["bestReview"] = sentAnal[0]

            gradeList = [
                int(instructors["grade_a_plus"]) + int(instructors["grade_a"]) + int(instructors["grade_a_minus"]),
                int(instructors["grade_b_plus"]) + int(instructors["grade_b"]) + int(instructors["grade_b_minus"]),
                int(instructors["grade_c_plus"]) + int(instructors["grade_c"]) + int(instructors["grade_c_minus"]),
                int(instructors["grade_d_plus"]) + int(instructors["grade_d"]) + int(instructors["grade_d_minus"]),
                int(instructors["grade_fail"])
            ]
            gradeListDetailed = [
                int(instructors["grade_a_plus"]) , int(instructors["grade_a"]) , int(instructors["grade_a_minus"]),
                int(instructors["grade_b_plus"]) , int(instructors["grade_b"]) , int(instructors["grade_b_minus"]),
                int(instructors["grade_c_plus"]) , int(instructors["grade_c"]) , int(instructors["grade_c_minus"]),
                int(instructors["grade_d_plus"]) + int(instructors["grade_d"]) + int(instructors["grade_d_minus"]),
                int(instructors["grade_fail"])
            ]
            courseforGradesSchema["gradeList"] = gradeList
            courseforGradesSchema["gradeListDetailed"] = gradeListDetailed
            

            if os.path.isfile(f"{DIR}/results/{item['code']}.csv"):
                tmp = []
                with open(f"{DIR}/results/{item['code']}.csv", "r") as f:
                    tmp = f.read()
                    tmp = tmp.split("\n")
                    tmp = tmp[:-1]
                

            facultySchema["sortedByGrades"].append(courseforGradesSchema)    
    for item in reviewsSortedByFaculty:
        if item["offer_dept"] == faculty:
            courseforReviewsSchema = {
                "courseCode" : item["code"],
                "courseName" : item["title"],
                "reviewRanges" : None,
                "allReviews" : None
            }
            sentAnal = []
            reviewRanges = [0,0,0,0,0]
            allReviews = []
            if os.path.isfile(f"{DIR}/results/{item['code']}.csv"):
                tmp = []
                with open(f"{DIR}/results/{item['code']}.csv", "r") as f:
                    tmp = f.read()
                    tmp = tmp.split("\n")
                    tmp = tmp[:-1]
                skipFirst = True
                for review in tmp:
                    if not skipFirst:
                        sentAnal.append(review.split(","))
                    skipFirst = False
                    
            if len(sentAnal) > 0:
                for review in sentAnal:
                    if len(review) >= 5:
                        if float(review[3]) > 0.8:
                            reviewRanges[0]+=1
                        elif float(review[3]) > 0.6:
                            reviewRanges[1]+=1
                        elif float(review[3]) > 0.4:
                            reviewRanges[2]+=1
                        elif float(review[3]) > 0.2:
                            reviewRanges[3]+=1
                        else:
                            reviewRanges[4]+=1
                        
                        rev = {
                            "actualReview" : review[4],
                            "positivityScore" : review[3]
                        }
                        allReviews.append(rev)
            courseforReviewsSchema["reviewRanges"] = reviewRanges
            courseforReviewsSchema["allReviews"] = allReviews
            facultySchema["sortedByReviews"].append(courseforReviewsSchema)

    allData.append(facultySchema)

with open(f"{DIR}schema.txt", "w") as f:
    f.write(json.dumps(allData))

                    
```
